```python
import pandas as pd
data={'height':[120,124,143,122,100,122,119],'weight':[60,72,67,76,78,69,90]}
df=pd.DataFrame(data)
print(df)
```

```python
x=df.weight
y=df.height
```

```python
import matplotlib.pyplot as plt    
```

```python
plt.scatter(x,y)
```

```python
xmean=x.mean()
xmean
```

```python
df['diffx']=xmean-x
df
```

```python
df["diffxsquare"]=df.diffx**2
df
```

```python
SSxx=df.diffxsquare.sum()
```

```python
ymean=y.mean()
```

```python
df['diffy']=ymean-y
df
```

```python
SSxy=(df.diffx*df.diffy).sum()
```

```python
m=SSxy/SSxx
m
```

```python
b=ymean-m*xmean
b
```

```python
def predict(value):
    pre=m*value+b
    return(pre)
```

```python
predict(60)
```

```python
plt.scatter(x,y)
plt.plot(x,m*x+b,'g')
```
