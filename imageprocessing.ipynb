```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
image=cv.imread ('./image.jpg')

gray_img = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

cv.imwrite ('newImage.jpg',gray_img)
cv.imshow ("new image grayscale",gray_img)
cv.waitKey (0)
cv.destroyAllWindows()
```

```python
Image_Height=gray_img.shape[0]
Image_Width=gray_img.shape[1]
Histogram=np.zeros([256],np.int32)
for x in range(0,Image_Height):
    for y in range(0,Image_Width):
        Histogram[gray_img[x,y]]+=1
print(Histogram)
```

```python
plt.figure()
plt.title("GrayScale Histogram")
plt.xlabel("Intensity Level")
plt.ylabel("Intensity Frequency")
plt.hist(gray_img.flat,edgecolor="black")
plt.show()
```

```python
from pickletools import uint8
uni_noise=np.zeros((2500,3200),dtype=np.uint8)
cv.randu(uni_noise,0,255)
uni_noise=(uni_noise*0.5).astype(np.uint8)
fig=plt.figure(dpi=300)
fig.add_subplot(1,3,2)
plt.imshow(uni_noise,cmap='gray')
plt.axis("off")
plt.title("Uniform Noise")
```

```python
plt.hist(uni_noise.flat,edgecolor="black")
plt.xlabel("noise pixel value");plt.ylabel("frequency")
plt.show()
```

```python
imagefornoise=cv.imread("./newImage.jpg",0)
print(imagefornoise.shape)
un_img=cv.add(imagefornoise,uni_noise)
fig.add_subplot(1,3,3)
plt.imshow(un_img,cmap='gray')
plt.axis("off")
plt.title("Combined")
```

```python
plt.hist(un_img.flat,edgecolor="black")
plt.show()
```
