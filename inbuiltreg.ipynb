```python
import pandas as pd
data=pd.read_csv("data.csv")
```

```python
x=data.sqft_living/max(data.sqft_living)
y=data.price/max(data.price)
```

```python
import matplotlib.pyplot as plt    
```

```python
plt.scatter(x,y)
```

```python
xmean=x.mean()
xmean
```

```python
data['diffx']=xmean-x
```

```python
data["diffxsquare"]=data.diffx**2
```

```python
SSxx=data.diffxsquare.sum()
```

```python
ymean=y.mean()
```

```python
data['diffy']=ymean-y
```

```python
SSxy=(data.diffx*data.diffy).sum()
```

```python
m=SSxy/SSxx

m
```

```python
b=ymean-m*xmean
b
```

```python
def predict(value):
    pre=m*value+b
    return(pre)
```

```python
plt.scatter(x,y)
plt.plot(x,m*x+b,'g')
```
