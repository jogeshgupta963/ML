```python
import pandas as pd
import matplotlib.pyplot as plt    
```

```python
data =pd.read_csv("./data.csv")
df = pd.DataFrame(data)
```

```python
y = df.price/max(df.price)
x = df.sqft_living/max(df.sqft_living)
plt.scatter(x,y)
```

```python
df["y_diff"]= y-y.mean()
df["x_diff"] = x-x.mean()
df["x_diff_square"] = x**2
```

```python
SSxx = df.x_diff_square.sum()
SSxy = (df.x_diff*df.y_diff).sum()
```

```python
m = SSxy/SSxx
b = y.mean() + m*x.mean()
print(m)
print(b)
```

```python
def predict(val):
    pre = m*val+b
    return pre
```
