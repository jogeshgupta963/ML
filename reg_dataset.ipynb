```python
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

data = pd.read_csv("./data.csv")
df = pd.DataFrame(data)

```

```python
# y = df.price
# x = df.sqft_living
x=data.sqft_living/max(data.sqft_living)
y=data.price/max(data.price)
plt.scatter(x,y)
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)
print(slope)
print(intercept)
plt.scatter(x,y)
plt.plot(x,slope*x+intercept,"g")
```
